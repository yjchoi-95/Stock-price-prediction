```python
feat = []
for li in feature_list:
    feat.extend(list(li))

select_idx = np.where(np.array(list(Counter(feat).values())) > 2)[0]
imp_feat = np.array(list(Counter(feat).keys()))[select_idx]
# np.save(data_path + "imp_feat.npy", imp_feat)
imp_feat2 = [re.sub("var", "close", x) for x in imp_feat]
imp_df = stock_df.loc[:, imp_feat2]
```

```python
def make_feat(x) :
    colname = re.sub("_close", "", x.name)
    
    # 볼린저 밴드의 경우 밴드 안에서의 위치값으로 변경
    upper, middle, lower = talib.BBANDS(x)
    bbands = ((x - lower) / (upper - lower))
    
    # 이평선의 경우 이평선 대비 현재 가격의 위치로 변경
    tema = x / talib.TEMA(x) 
    sma = x / talib.SMA(x)
    apo = talib.APO(x, fastperiod=12, slowperiod=26, matype=0)
    
    # macd hist만 사용해도될듯
    macd, macdsignal, macdhist = talib.MACDFIX(x, signalperiod=9)
    rsi = talib.RSI(x)
    
    # statistic indicator
    linear = x / talib.LINEARREG(x)
    tsf = x / talib.TSF(x)
    
    # cycle indicator
    trendmode = talib.HT_TRENDMODE(x)
    
    feat_name = ['bands','tema','sma','apo','macdhist','rsi','linear','tsf','trend']
    col_name = [colname+"_"+x for x in feat_name]
    
    output_df = pd.DataFrame([bbands, tema, sma, apo, macdhist, rsi, linear, tsf, trendmode]).T
    output_df.columns = col_name
    
    return output_df
```

```python
for idx, cols in enumerate(imp_df.columns):
    if idx == 0:
        feature_df = make_feat(imp_df[cols])
    else :
        feature_df = pd.concat([feature_df, make_feat(imp_df[cols])], axis = 1)
```
